# Display 3-D images on a grid

```python
import seaborn_image as isns
from skimage import exposure # for adjusting image exposure
```

Load a multi-dimension image array

```python
cells = isns.load_image("cells")
```

```python
cells.shape
```

```python
g = isns.ImageGrid(cells, slices=[10,20,30,40,50,55])
```

```python
g = isns.ImageGrid(cells, slices=[10,20,30,40,50,55], cbar=False)
```

```python
g = isns.ImageGrid(cells, cbar=False, height=1, col_wrap=10)
```

```python
cells_low_gamma = exposure.adjust_gamma(cells, gamma=0.5)

g = isns.ImageGrid(cells_low_gamma, cbar=False, height=1, col_wrap=10)
```
